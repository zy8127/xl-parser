```python
import pandas as pd
import sqlite3

con = sqlite3.connect('db.sqlite3')
df = pd.read_sql(sql='select * from tc拉拔转序', con=con)
df = df.set_index(['产品图号', '工序'])
del df['产品类别']
del df['合计']

df = df.stack()

df.index.names = ['产品图号', '工序', '日期']
df = df.to_frame()
df.columns = ['数量']

df.to_sql(name='test', con=con, flavor='sqlite', if_exists='replace')
df
```
